# Using Dropouts

Dropout is a generealization/regularization system to reduce the overfitting. In this project, we will look into that.

```python
device = "cuda"
```

```python
import torch
import torch.nn.functional as F
from torchvision import datasets, transforms
from tqdm.notebook import tqdm
from matplotlib import pyplot as plt
```

## Load Data

```python
transform_pil=transforms.Compose([
  transforms.ToTensor(), # first, convert image to PyTorch tensor
  transforms.Normalize((0.1307,), (0.3081,)) # normalize inputs
])
```

```python
ds_train = datasets.FashionMNIST("~/data", train=True, download=True, transform=transform_pil)
ds_valid = datasets.FashionMNIST("~/data", train=True, download=True, transform=transform_pil)
```

```python
_dl = torch.utils.data.DataLoader(ds_train, shuffle=True, batch_size=16)
sample_batch = next(iter(_dl))
I.wrap(*sample_batch).show(cols=8)
```

## Model & Training

```python
epochs = 10
bs = 16
lr = 0.1
wd = 0.1

dl_train = torch.utils.data.DataLoader(ds_train, shuffle=True, batch_size=16)

w0 = torch.randn(28*28, 256, requires_grad=True, device=device).float()
b0 = torch.randn(256, requires_grad=True, device=device).float()
w1 = torch.randn(256, 10, requires_grad=True, device=device).float()
b1 = torch.randn(10, requires_grad=True, device=device).float()

loss_list = []
valid_loss_list = []
accuracy_list = []
valid_accuracy_list = []

def forward(X):
    h = torch.relu(X@w0 + b0)
    y = h@w1 + b1
    return y

def calc_accuracy(images, labels):
    batch_size = len(images)
    pred = forward(images.reshape(batch_size, -1))
    pred = pred.detach()
    
    pred_labels = torch.argmax(pred, dim=-1)
    accuracy = (pred_labels == labels).sum() / batch_size * 100
    
    return accuracy

lp = I.live_plot(2)
i = 0

for epoch in range(epochs):
    for batch in tqdm(dl_train):
        i += 1;
        images = batch[0].to(device)
        labels = batch[1].to(device)
        batch_size = len(images)

        if w0.grad is not None:
            w0.grad.zero_()
            b0.grad.zero_()
            w1.grad.zero_()
            b1.grad.zero_()

        yhat = forward(images.reshape(batch_size, -1))
        l2_loss = (torch.norm(w0) + torch.norm(w1))/2 * wd
        loss = F.cross_entropy(yhat, labels) + l2_loss

        loss.backward()

        with torch.no_grad():
            w0 -= lr * w0.grad
            b0 -= lr * b0.grad
            w1 -= lr * w1.grad
            b1 -= lr * b1.grad

        if i %100 == 0:
            with torch.no_grad():
                accuracy_list.append(calc_accuracy(images, labels).cpu().item())
                loss_list.append(loss.item())
                
                _dl = torch.utils.data.DataLoader(ds_train, shuffle=True, batch_size=bs)
                sample = next(iter(_dl))
                valid_images, valid_lables = sample[0].to(device), sample[1].to(device)
                
                valid_accuracy_list.append(calc_accuracy(valid_images, valid_lables).cpu().item())
                valid_loss = F.cross_entropy(forward(valid_images.reshape(bs, -1)), valid_lables) + l2_loss
                valid_loss_list.append(valid_loss.item())
                
            lp.plot(0, loss_list, label="train loss")
            lp.plot(0, valid_loss_list, label="valid loss")
            lp.plot(1, accuracy_list, label="train accuracy")
            lp.plot(1, valid_accuracy_list, label="valid accuracy")
            lp.update()
        
lp.close()
```

```python
#Let's check the total accuracy

_dl = torch.utils.data.DataLoader(ds_valid, shuffle=True, batch_size=128)
accuracy_list = []
for batch in tqdm(_dl):
    images = batch[0].to(device)
    labels = batch[1].to(device)
    
    with torch.no_grad():
        accuracy_list.append(calc_accuracy(images, labels))
    
torch.tensor(accuracy_list).mean().item()
```

```python
def dropout(W, ratio):
    if ratio == 1.0:
        return W * 0
    
    mask = torch.rand_like(W) > ratio
    return (mask * W).float() / (1-ratio)

dropout(WW, 0.5)
```

## The Loop with the DropOut

```python
epochs = 10
bs = 16
lr = 0.1
wd = 0.1

dl_train = torch.utils.data.DataLoader(ds_train, shuffle=True, batch_size=16)

w0 = torch.randn(28*28, 256, requires_grad=True, device=device).float()
b0 = torch.randn(256, requires_grad=True, device=device).float()
w1 = torch.randn(256, 10, requires_grad=True, device=device).float()
b1 = torch.randn(10, requires_grad=True, device=device).float()

loss_list = []
valid_loss_list = []
accuracy_list = []
valid_accuracy_list = []

def forward(X):
    h = torch.relu(X@w0 + b0)
    h = dropout(h, 0.01)
    y = h@w1 + b1
    return y

def calc_accuracy(images, labels):
    batch_size = len(images)
    pred = forward(images.reshape(batch_size, -1))
    pred = pred.detach()
    
    pred_labels = torch.argmax(pred, dim=-1)
    accuracy = (pred_labels == labels).sum() / batch_size * 100
    
    return accuracy

lp = I.live_plot(2)
i = 0

for epoch in range(epochs):
    for batch in tqdm(dl_train):
        i += 1;
        images = batch[0].to(device)
        labels = batch[1].to(device)
        batch_size = len(images)

        if w0.grad is not None:
            w0.grad.zero_()
            b0.grad.zero_()
            w1.grad.zero_()
            b1.grad.zero_()

        yhat = forward(images.reshape(batch_size, -1))
        l2_loss = (torch.norm(w0) + torch.norm(w1))/2 * wd
        loss = F.cross_entropy(yhat, labels) + l2_loss

        loss.backward()

        with torch.no_grad():
            w0 -= lr * w0.grad
            b0 -= lr * b0.grad
            w1 -= lr * w1.grad
            b1 -= lr * b1.grad

        if i %100 == 0:
            with torch.no_grad():
                accuracy_list.append(calc_accuracy(images, labels).cpu().item())
                loss_list.append(loss.item())
                
                _dl = torch.utils.data.DataLoader(ds_train, shuffle=True, batch_size=bs)
                sample = next(iter(_dl))
                valid_images, valid_lables = sample[0].to(device), sample[1].to(device)
                
                valid_accuracy_list.append(calc_accuracy(valid_images, valid_lables).cpu().item())
                valid_loss = F.cross_entropy(forward(valid_images.reshape(bs, -1)), valid_lables) + l2_loss
                valid_loss_list.append(valid_loss.item())
                
            lp.plot(0, loss_list, label="train loss")
            lp.plot(0, valid_loss_list, label="valid loss")
            lp.plot(1, accuracy_list, label="train accuracy")
            lp.plot(1, valid_accuracy_list, label="valid accuracy")
            lp.update()
        
lp.close()
```

```python
#Let's check the total accuracy

_dl = torch.utils.data.DataLoader(ds_valid, shuffle=True, batch_size=128)
accuracy_list = []
for batch in tqdm(_dl):
    images = batch[0].to(device)
    labels = batch[1].to(device)
    
    with torch.no_grad():
        accuracy_list.append(calc_accuracy(images, labels))
    
torch.tensor(accuracy_list).mean().item()
```
