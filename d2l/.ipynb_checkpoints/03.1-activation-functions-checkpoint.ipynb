# Various Activation Functions

We talk about RelU, Sigmoid & TanH

```python
import torch
from matplotlib import pyplot as plt
```

## Sigmoid

This is a function which gives us 0-1 range for anything it accepts. So, this is pretty useful as the last layer for probability related models.

```python
X = torch.arange(-8, 8, 0.1)
X.requires_grad_()
Y = torch.sigmoid(X)
Y.backward(torch.ones_like(X))
```

```python
plt.plot(X.detach().cpu(), Y.detach().cpu(), label="Sigmoid Value")
plt.plot(X.detach().cpu(), X.grad, label="Gradient")
plt.legend()
```

**As you can see it has a range between 0-1. But after 5 and -5 it looks like values will be scattered at the edges.**

Also the gradients will be minimum at the edges. This might cause models to not learn even we increase the learning rate.

## Tanh

A function which gives use -1 to 1 range for any value it take.

```python
X = torch.arange(-8, 8, 0.1)
X.requires_grad_()
Y = torch.tanh(X)
Y.backward(torch.ones_like(X))
```

```python
plt.plot(X.detach().cpu(), Y.detach().cpu(), label="Tanh value")
plt.plot(X.detach().cpu(), X.grad, label="Gradient")
```

**Most of the properties are similar to Sigmoid but the value range is between -1 and 1.**

## ReLU

This is a pretty simple function, but it has ver cool properties.

```python
X = torch.arange(-8, 8, 0.1)
X.requires_grad_()
Y = torch.relu(X)
Y.backward(torch.ones_like(X))
```

```python
plt.plot(X.detach().cpu(), Y.detach().cpu(), label="ReLU value")
plt.plot(X.detach().cpu(), X.grad, label="Gradient")
```

**As you can see, it simply the Max(X, 0) function which always set 0 for any negative values**

But have a look at the gradient. It's always 1 for any positive value. That's a good sign for making a good learnable layer.

```python
## See. this is how you define ReLU.
X = torch.arange(-8, 8, 0.1)
X.requires_grad_()
Y = torch.maximum(X, torch.tensor(0))
Y.backward(torch.ones_like(X))

plt.plot(X.detach().cpu(), Y.detach().cpu(), label="ReLU value")
plt.plot(X.detach().cpu(), X.grad, label="Gradient")
```

## Leaky ReLU

This is a slightly modified version of the ReLU.

```python
## See. this is how you define ReLU.
X = torch.arange(-8, 2, 0.1)
X.requires_grad_()
Y = torch.nn.functional.leaky_relu(X, negative_slope=0.25)
Y.backward(torch.ones_like(X))

plt.plot(X.detach().cpu(), torch.zeros_like(X))
plt.plot(X.detach().cpu(), Y.detach().cpu(), label="ReLU value")
plt.plot(X.detach().cpu(), X.grad, label="Gradient")
plt.legend()
```

As you can see, it will leak some values on the negative side. Basically multiply the values by some factor.

In this case, we used a larger value of `0.25`, but it can be `0.01` or lower.

**This is very important as it gives us non-zero gradient for the negative side.**
