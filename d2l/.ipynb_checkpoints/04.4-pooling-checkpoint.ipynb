## Core Functions

```python
def apply_pad(tensor, n_pad=1):
    tw, th = tensor.shape
    y = torch.zeros(tw + n_pad * 2, th + n_pad *2).to(tensor.device)
    y[n_pad:n_pad + tw, n_pad:n_pad + th] = tensor
    return y
```

```python
def run_pool(X, pool_size, n_pad=0, n_stride=1):
    assert len(X.shape) == 2, "Only 2D tensors are supported"
    X = apply_pad(X, n_pad)
    kw, kh = pool_size, pool_size
    xw, xh = X.shape
    yw = math.floor((xw - kw) / n_stride  + 1)
    yh = math.floor((xh - kh) / n_stride  + 1)
    
    Y = torch.zeros(yw, yh).to(X.device)
    for u in range(yw):
        for v in range(yh):
            xu, xv = u*n_stride, v*n_stride
            selection = X[xu:xu+kw, xv:xv+kh]
            Y[u, v] = (selection).mean()
    return Y
```

```python
%time im = run_pool(image[0][0], 8, n_stride=4)
im.shape, I.wrap(im).show()
```

## Use A Faster Version

We can use a faster version which comes with Pytorch instead of this.

```python
pool_layer = torch.nn.AvgPool2d(kernel_size=8, stride=4)
%time im2 = pool_layer(image[0])
I.wrap(im2[0]).show()
```
