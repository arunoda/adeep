```python
## Load the Word Embeddings
import pickle
import torch
import torch.nn as nn
from pathlib import Path
from lib.glove import GloveEmbeddings
```

```python
glove_pkl_path = Path.home()/"data"/"glove.pkl"
glove = pickle.load(open(glove_pkl_path, "rb"))
```

```python
glove.make(source_dataset[22313]['text']).shape
```

```python
class TwitterDataset(torch.utils.data.Dataset):
    def __init__(self, source):
        super().__init__()
        self.source = source
        
    def __len__(self):
        return len(self.source)
    
    def __getitem__(self, index):
        raw_item = self.source[index]
        item = glove.make(raw_item['text'])[0:10, :]
        zeros = torch.zeros(10, 50)
        zeros[0:len(item)] = item
        
        return zeros, raw_item['sentiment']
```

```python
train_dataset = TwitterDataset(source_dataset)
```

```python
test_inputs = next(iter(torch.utils.data.DataLoader(train_dataset, batch_size=3, shuffle=True)))[0]
test_inputs.shape
```

## Training with a Liner Model (Baseline)

```python
device = "cuda"
```

```python
from tqdm.notebook import tqdm
```

```python
class BaseLineModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(50, 5),
            nn.ReLU()
        )
        
    def forward(self, inputs):
        result = self.net(inputs)
        result = torch.mean(result, dim=1)
        return result
    
_m = BaseLineModel()
_m(test_inputs).shape
```

### The Loop

```python
train_dl = torch.utils.data.DataLoader(train_dataset, shuffle=True, batch_size=16)
model = BaseLineModel().to(device)
optimizer = torch.optim.Adam(model.parameters(), lr=0.05)
calc_loss = torch.nn.CrossEntropyLoss()

lp = I.live_plot(2)
i = 0

losses = []

for batch in tqdm(train_dl):
    optimizer.zero_grad()
    
    inputs, labels = batch[0].to(device), batch[1].to(device)
    preds = model(inputs)
    loss = calc_loss(preds, labels)
    
    with torch.no_grad():
        losses.append(loss.item())
        loss.backward()
        optimizer.step()
        
    i += 1
    if i % 10 == 0:
        lp.plot(0, losses)
        lp.update()
        
    if i == 500:
        break

lp.close()
```

## Now Let's do the same with an Attention based Model

```python
test_inputs.shape
```

```python
class AttentionModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.att = nn.MultiheadAttention(embed_dim=50, num_heads=2)
        self.ff = nn.Linear(50, 5)
        
    def forward(self, inputs):
        result, _ = self.att(inputs, inputs, inputs)
        result = nn.functional.relu(result)
        result = self.ff(result)
        result = nn.functional.relu(result)
        result = torch.mean(result, dim=1)
        return result
    
_m = AttentionModel()
_m(test_inputs).shape
```

### The Loop

```python
train_dl = torch.utils.data.DataLoader(train_dataset, shuffle=True, batch_size=16)
model = AttentionModel().to(device)
optimizer = torch.optim.Adam(model.parameters(), lr=0.05)
calc_loss = torch.nn.CrossEntropyLoss()

lp = I.live_plot(2)
i = 0

losses = []

for batch in tqdm(train_dl):
    optimizer.zero_grad()
    
    inputs, labels = batch[0].to(device), batch[1].to(device)
    preds = model(inputs)
    loss = calc_loss(preds, labels)
    
    with torch.no_grad():
        losses.append(loss.item())
        loss.backward()
        optimizer.step()
        
    i += 1
    if i % 10 == 0:
        lp.plot(0, losses)
        lp.update()
        
    if i == 500:
        break

lp.close()
```

**Well. It converges, quickly than the Linear model**

## Let's use a Transformer Block

```python
class TransformerBlock(nn.Module):
    def __init__(self):
        super().__init__()
        self.att = nn.MultiheadAttention(embed_dim=50, num_heads=5)
        self.norm1 = nn.LayerNorm(50)
        self.norm2 = nn.LayerNorm(50)
        self.ff = nn.Sequential(
            nn.Linear(50, 200),
            nn.ReLU(),
            nn.Linear(200, 50)
        )
        
    def forward(self, inputs):
        y, _ = self.att(inputs, inputs, inputs) #attention
        y = self.norm1(y) # normalization
        y = inputs + y # residual (only each embeddings)
        
        y = self.ff(y) # feed forward network
        y = inputs + y # residual
        y = self.norm2(y) # normalization (only each embeddings)
        
        return y
    
_m = TransformerBlock()
_m(test_inputs).shape
```

```python
class ClassicationTransformer(nn.Module):
    def __init__(self):
        super().__init__()
        self.tranformers = nn.Sequential(
            TransformerBlock(),
            TransformerBlock(),
            TransformerBlock()
        )
        self.ff = nn.Sequential(
            nn.Linear(50, 5),
            nn.ReLU()
        )
        
    def forward(self, inputs):
        y = self.tranformers(inputs)
        y = torch.mean(y, dim=1)
        y = self.ff(y)
        return y
    
_m = ClassicationTransformer()
_m(test_inputs).shape
```

### Train Loop

```python
train_dl = torch.utils.data.DataLoader(train_dataset, shuffle=True, batch_size=16)
model = ClassicationTransformer().to(device)
optimizer = torch.optim.Adam(model.parameters(), lr=0.05)
calc_loss = torch.nn.CrossEntropyLoss()

lp = I.live_plot(2)
i = 0

losses = []

for batch in tqdm(train_dl):
    optimizer.zero_grad()
    
    inputs, labels = batch[0].to(device), batch[1].to(device)
    preds = model(inputs)
    loss = calc_loss(preds, labels)
    
    with torch.no_grad():
        losses.append(loss.item())
        loss.backward()
        optimizer.step()
        
    i += 1
    if i % 10 == 0:
        lp.plot(0, losses)
        lp.update()
        
    if i == 500:
        break

lp.close()
```
