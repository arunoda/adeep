```python
from PIL import Image
from fastcore.all import concat
import torch, logging
from pathlib import Path
from huggingface_hub import notebook_login
from diffusers import StableDiffusionPipeline
from PIL import Image
import matplotlib.pyplot as plt
logging.disable(logging.WARNING)
torch.manual_seed(1)

def image_grid(imgs, rows, cols):
    w,h = imgs[0].size
    grid = Image.new('RGB', size=(cols*w, rows*h))
    for i, img in enumerate(imgs): grid.paste(img, box=(i%cols*w, i//cols*h))
    return grid

def image_grid_flip(imgs, rows, cols):
    w,h = imgs[0].size
    grid = Image.new('RGB', size=(cols*w, rows*h))
    for i, img in enumerate(imgs):
        row = i % rows
        col = (int)((i - row) / rows)
        grid.paste(img, box=(w*col, h*row))
    return grid

def image_resize(grid, factor=3): 
    return grid.resize(((int)(grid.size[0]/factor), (int)(grid.size[1]/factor)))
```

```python
if not (Path.home()/'.huggingface'/'token').exists(): notebook_login()
```

## Load the Model

```python
pipe = StableDiffusionPipeline.from_pretrained("CompVis/stable-diffusion-v1-4", revision="fp16", torch_dtype=torch.float16) 
pipe = pipe.to("cuda")
```

```python
# Download some pretrained embeddings

from fastdownload import FastDownload
embeds_url = "https://huggingface.co/sd-concepts-library/indian-watercolor-portraits/resolve/main/learned_embeds.bin"
embeds_path = FastDownload().download(embeds_url)
embeds_dict = torch.load(str(embeds_path), map_location="cpu")
```

```python
tokenizer = pipe.tokenizer
text_encoder = pipe.text_encoder
new_token, embeds = next(iter(embeds_dict.items()))
embeds = embeds.to(text_encoder.dtype)
```

```python
new_token
```

```python
assert tokenizer.add_tokens(new_token) == 1, "The token already exists!"
```

```python
# assign new token's embdeddings
text_encoder.resize_token_embeddings(len(tokenizer))
new_token_id = tokenizer.convert_tokens_to_ids(new_token)
text_encoder.get_input_embeddings().weight.data[new_token_id] = embeds
```

## Let's Play with this

```python
torch.manual_seed(1000)
image = pipe("Woman reading in the style of <watercolor-portrait>").images[0]
image_resize(image, 2)
```

```python
torch.manual_seed(1000)
images = pipe(["dog reading a book in the style of <watercolor-portrait>"] * 5).images
```

```python
image_resize(image_grid(images, 1, 5), 2)
```
