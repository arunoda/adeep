```python
from PIL import Image
from fastcore.all import concat
import torch, logging
from pathlib import Path
from huggingface_hub import notebook_login
from diffusers import StableDiffusionPipeline
from PIL import Image
import matplotlib.pyplot as plt
logging.disable(logging.WARNING)
torch.manual_seed(1)

def image_grid(imgs, rows, cols):
    w,h = imgs[0].size
    grid = Image.new('RGB', size=(cols*w, rows*h))
    for i, img in enumerate(imgs): grid.paste(img, box=(i%cols*w, i//cols*h))
    return grid

def image_grid_flip(imgs, rows, cols):
    w,h = imgs[0].size
    grid = Image.new('RGB', size=(cols*w, rows*h))
    for i, img in enumerate(imgs):
        row = i % rows
        col = (int)((i - row) / rows)
        grid.paste(img, box=(w*col, h*row))
    return grid

def image_resize(grid, factor=3): 
    return grid.resize(((int)(grid.size[0]/factor), (int)(grid.size[1]/factor)))
```

## Hello World

```python
pipe = StableDiffusionPipeline.from_pretrained("CompVis/stable-diffusion-v1-4", revision="fp16", torch_dtype=torch.float16).to("cuda")
```

```python
prompt = "a photograph of an astronaut riding a horse"
```

```python
image_resize(pipe(prompt).images[0], 2)
```

```python
## With a Seed
torch.manual_seed(2223)
image_resize(pipe(prompt).images[0], 2)
```

## Steps

Here, we use steps to let it generate some decent image. 50 seems like the good solution.
But we can of course change it. Let's give it a try.

```python
#let do it for couple of steps
images = concat(pipe(prompt, num_inference_steps=i).images for i in [1, 3, 16, 30, 50, 100])
```

```python
grid = image_grid(images, 1, 6)
image_resize(grid, 3)
```

**As you can see, there not much difference between 50 & 100. Even the difference between 30 & 50 is also minimum.**

*It seems like 50 is the golden standard*

## Guidance Scale

This is basically about using the given propmt or not. 

* A lower values means, not really looking at the prompt.
* Higer values means, it looking at the prompt a lot.

```python
images = concat(pipe([prompt]*6, guidance_scale=i).images for i in [1, 4, 7, 15, 20])
```

```python
grid = image_grid(images, 5, 6)
image_resize(grid, 4)
```

That's interesting. (Check rows)

* 1 is totally BS
* 4 & 7 gives us interesting results
* 15 & 20 follow the prompt, but they are bit sketchy & may be not in high quality

## Negative Prompts

Here, we use a negative prompt to get rid of something inside the picture.

```python
torch.manual_seed(3233)
prompt2 = "A dog walking on the water, 4K high def"
image_resize(pipe(prompt2).images[0], 2)
```

```python
## Let's remove grass

torch.manual_seed(3233)
image_resize(pipe(prompt2, negative_prompt="grass").images[0], 2)
```

```python
## lets remove both grass and trees

torch.manual_seed(3233)
image_resize(pipe(prompt2, negative_prompt="grass, mountain, tree, green, beach").images[0], 2)
```

**As you can see, it's a bit hard to get rid of trees at the distance.**
