```python
import torch
from torchvision import transforms
from tqdm.notebook import tqdm # A progress bar library
from matplotlib import pyplot as plt
import torch.nn.functional as F
```

```python
# loading CLIP
from CLIP import clip # The clip model
perceptor = clip.load('ViT-B/32', jit=False)[0].eval().requires_grad_(False).to(device)
```

```python
# Encoding some texts
prompt = "A watercolor painting of the sunset over a mountain range"
text_embs = perceptor.encode_text(clip.tokenize(prompt).to(device))
text_embs.shape
```

```python
# Checking the image resolution it needs
perceptor.visual.input_resolution
```

```python
# encoding an image
torch.manual_seed(23)
im = torch.randn(1, 3, 244, 244).to(device)
normalize = transforms.Normalize(mean=[0.48145466, 0.4578275, 0.40821073],std=[0.26862954, 0.26130258, 0.27577711])
img_embs = perceptor.encode_image(normalize(im).to(device))
img_embs.shape
```

```python
## See. Both of those contains the same shape
```

```python
# Creating a loss function
def calc_loss(text_embs, im):
    img_embs = perceptor.encode_image(normalize(im))
    error = text_embs - img_embs
    loss = (error ** 2).mean()**0.5
    return loss
```

```python
calc_loss(text_embs, im)
```

```python
# let's start training
prompt = "A watercolor painting of the sunset over a mountain range"
text_embs = perceptor.encode_text(clip.tokenize(prompt).to(device))

torch.manual_seed(23)
im = torch.rand(1, 3, 244, 244)
original_im = im.clone()
im = im.to(device)
im.requires_grad_()

optimizer = torch.optim.SGD([im], lr=500)
losses = []

for i in tqdm(range(200)):
    # grad zero
    optimizer.zero_grad()
    
    # calc loss
    loss = calc_loss(text_embs, im)
    losses.append(loss.item())
    
    # updating params
    loss.backward()
    optimizer.step()
    
fig, ax = plt.subplots(1, 4, figsize=(15, 3))
ax[0].plot(losses)
ax[1].imshow(original_im.squeeze(0).permute(1, 2, 0))
ax[2].imshow(im.detach().cpu().squeeze(0).permute(1, 2, 0))
ax[3].imshow((im.detach().cpu() - original_im).squeeze(0).permute(1, 2, 0))
```

**See. There's not much different over here.**

Here are some issues:

* Image data clipping
* Loss function is bad for this
* We need to transform the image in various ways before calculating the loss (otherwise, it will update the image as boxes)

```python
# The new loss function
tfms = transforms.Compose([
    transforms.RandomResizedCrop(224),
    transforms.RandomAffine(5),
    transforms.ColorJitter(),
    transforms.GaussianBlur(5),
])

# Our new and improved loss function
def calc_loss2(text_embed, im):
    im = im.clip(0, 1)
    image_embed = perceptor.encode_image(tfms(im).to(device)).float()
    input_normed = F.normalize(image_embed.unsqueeze(1), dim=2)
    embed_normed = F.normalize(text_embed.unsqueeze(0), dim=2)
    dists = input_normed.sub(embed_normed).norm(dim=2).div(2).arcsin().pow(2).mul(2) # Squared Great Circle Distance
    return dists.mean()
```

```python
# Let's train again
prompt = "A watercolor painting of the sunset over a mountain range"
text_embs = perceptor.encode_text(clip.tokenize(prompt).to(device))

torch.manual_seed(23)
im = torch.rand(1, 3, 244, 244)
original_im = im.clone()
im = im.to(device)
im.requires_grad_()

optimizer = torch.optim.SGD([im], lr=500)
losses = []

for i in tqdm(range(200)):
    # grad zero
    optimizer.zero_grad()
    
    # calc loss
    loss = calc_loss2(text_embs, im)
    losses.append(loss.item())
    
    # updating params
    loss.backward()
    optimizer.step()
    
fig, ax = plt.subplots(1, 4, figsize=(15, 3))
ax[0].plot(losses)
ax[1].imshow(original_im.squeeze(0).permute(1, 2, 0))
ax[2].imshow(im.detach().cpu().squeeze(0).permute(1, 2, 0))
ax[3].imshow((im.detach().cpu() - original_im).squeeze(0).permute(1, 2, 0))
```

**It's looking bit okay now**

```python
## Shall we use ADAM
# Let's train again
prompt = "A watercolor painting of the sunset over a mountain range"
text_embs = perceptor.encode_text(clip.tokenize(prompt).to(device))

torch.manual_seed(23)
im = torch.rand(1, 3, 244, 244)
original_im = im.clone()
im = im.to(device)
im.requires_grad_()

optimizer = torch.optim.Adam([im], lr=0.05)
losses = []

for i in tqdm(range(200)):
    # grad zero
    optimizer.zero_grad()
    
    # calc loss
    loss = calc_loss2(text_embs, im)
    losses.append(loss.item())
    
    # updating params
    loss.backward()
    optimizer.step()
    
fig, ax = plt.subplots(1, 4, figsize=(15, 3))
ax[0].plot(losses)
ax[1].imshow(original_im.squeeze(0).permute(1, 2, 0))
ax[2].imshow(im.detach().cpu().squeeze(0).permute(1, 2, 0))
ax[3].imshow((im.detach().cpu() - original_im).squeeze(0).permute(1, 2, 0))
```
