## Simple Convolutions based VAE

```python
import torch
import torch.nn as nn
from torchvision import datasets, transforms
from matplotlib import pyplot as plt
from tqdm.notebook import tqdm
```

```python
ds_train = datasets.MNIST('~/.data/MNIST', download=True, train=True, transform=transforms.ToTensor())
```

```python
sample_image = ds_train[0][0].unsqueeze(0)
sample_image.shape
```

```python
class Encoder(nn.Module):
    def __init__(self):
        super().__init__()
        
        self.layers = nn.Sequential(
            nn.Conv2d(1, 32, kernel_size=3),
            nn.ReLU(),
            nn.Conv2d(32, 64, kernel_size=3),
            nn.ReLU(),
            nn.Conv2d(64, 96, kernel_size=4, stride=2),
            nn.ReLU(),
            nn.Flatten(),
            nn.Linear(96*11*11, 1024),
            nn.ReLU(),
            nn.Linear(1024, 64)
        )
        
    def forward(self, inputs):
        return self.layers(inputs)
    
_enc = Encoder()
_enc(sample_image).shape
```

```python
class Decoder(nn.Module):
    def __init__(self):
        super().__init__()
        
        self.layers = nn.Sequential(
            nn.Linear(64, 1024),
            nn.ReLU(),
            nn.Linear(1024, 96*11*11),
            nn.ReLU(),
            nn.Unflatten(dim=-1, unflattened_size=(96, 11, 11)),
            nn.ConvTranspose2d(96, 64, 4, stride=2),
            nn.ReLU(),
            nn.ConvTranspose2d(64, 32, 3),
            nn.ReLU(),
            nn.ConvTranspose2d(32, 1, 3),
        )
        
    def forward(self, inputs):
        return self.layers(inputs)
    
_dec = Decoder()
_decoded = _dec(_enc(sample_image))
print(_decoded.shape)
plt.imshow(_decoded[0][0].detach().cpu())
```

## Training

```python
device = "cuda"
```

```python
torch.manual_seed(23)

enc = Encoder().to(device)
dec = Decoder().to(device)

optimizer_enc = torch.optim.Adam(enc.parameters(), lr=0.005, weight_decay=0.0001)
optimizer_dec = torch.optim.Adam(dec.parameters(), lr=0.005, weight_decay=0.0001)

calc_loss = torch.nn.MSELoss()

train_dl = torch.utils.data.DataLoader(ds_train, batch_size=32, shuffle=True)

losses = []

for epoch in range(4):
    losses = []
    for images, _ in tqdm(train_dl):
        images = images.to(device)

        optimizer_enc.zero_grad()
        optimizer_dec.zero_grad()

        encoded = enc(images)
        decoded = dec(encoded)

        batch_size = len(images)
        loss = calc_loss(decoded.reshape(batch_size, -1), images.reshape(batch_size, -1))
        losses.append(loss.item())
        loss.backward()

        optimizer_enc.step()
        optimizer_dec.step()
    
    print(f"Loss: {torch.tensor(losses).mean()}")
```

## Evaluations

```python
ds_eval = datasets.MNIST('~/.data/MNIST', download=True, train=False, transform=transforms.ToTensor())
```

```python
with torch.no_grad():
    _, ax = plt.subplots(2, 20, figsize=(20, 2))
    for i in range(20):
        image = ds_eval[i][0].to(device)
        encoded = dec(enc(image.unsqueeze(0)))

        ax[0][i].imshow(image.cpu()[0], cmap="gray")
        ax[0][i].axis("off")
        ax[1][i].imshow(encoded.cpu()[0][0], cmap="gray")
        ax[1][i].axis("off")
            
```

**It's nice to see that we can get a result like this, with some very little training**

## Generation with Random Values

```python
torch.manual_seed(43)
source = torch.randn(10, 1, 28, 28);
with torch.no_grad():
    _, ax = plt.subplots(2, 10, figsize=(20, 4))
    encoded = dec(enc(source.to(device)))
    
    for i in range(10):
        ax[0][i].imshow(source.cpu()[i][0], cmap="gray")
        ax[0][i].axis("off")
        ax[1][i].imshow(encoded.cpu()[i][0], cmap="gray")
        ax[1][i].axis("off")
```
