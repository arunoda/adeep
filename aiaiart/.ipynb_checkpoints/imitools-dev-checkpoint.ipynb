```python
import torch
import imitools as I
```

```python
pt_images = torch.rand(100, 3, 8, 8)
I.wrap(pt_images).cpil().resize(256, resample=4).to_video().show()
```

```python
I.wrap(pt_images).show(max_count=100, scale=1, cols=20, captions=False)
```

```python
images = I.wrap(pt_images)
```

```python
from PIL import Image
```

```python
images.cpil().resize(256, resample=Image.BOX).pil()[1]
```
